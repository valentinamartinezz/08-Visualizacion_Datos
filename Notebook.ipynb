## Variables.


 - `age`= La edad del individuo.          
 - ``gender`` = El género del individuo, ya sea ``Male`` - Hombre o ``Female`` - Mujer.           
 - ``occupation`` = La ocupación del individuo.     
 - ``education_level`` = El nivel más alto de educación alcanzado por el individuo, como 'Escuela secundaria', 'Asociado', 'Licenciatura', 'Maestría' o 'Doctorado'. 
 - ``marital_status`` = El estado civil de la persona, ya sea "Soltero" o "Casado".
 - ``income`` = El ingreso anual individuo en dólares.          
 - ``credit_score`` = El puntaje de crédito de la persona, que va de 300 a 850.   
 - ``loan_status`` = La variable de destino, que indica si la solicitud de préstamo fue "Aprobada" o "Denegada".

# 2.Crear un grafico de barras en donde se muestren la cantidad de personas registradas por ocupació.

```python
prest['occupation'].value_counts()
```

```python
a=prest['occupation'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.bar(a.index,a.values,color='red')
plt.title('Profesiones de los clientes')
plt.xlabel('Profeciones')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```

# 3.Crea un gráfico de barras en donde se muestre el total de personar registradas por education_level.

```python
a=prest['education_level'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.bar(a.index,a.values,color='black')
plt.title('Nivel Educativo')
plt.xlabel('Profeciones')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```

# 4.Crea un grafico circular en donde se muestre la proporcion de Hombres y mujeres de acuerdo a la columna gender.

```python
a=prest['gender'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.pie(a.values, labels=a.index, colors=['red', 'blue'], autopct='%1.1f%%')

#añadir título y  etiquetas a los ejes.
plt.title('Proporción de hombres y mujeres')
plt.xlabel('Generos')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```

# 5.Crea un histograma con la columna age.

```python
a=prest['age'].value_counts()
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
plt.hist(prest['age'], bins=10, color='red', edgecolor='black')  # Crear histograma

# Añadir título y etiquetas a los ejes.
plt.title('Distribución de Edad')
plt.xlabel('Edad')  # Poner etiqueta al eje x
plt.ylabel('Número de Clientes')  # Poner etiqueta al eje y
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x si es necesario
plt.show()  # Mostrar gráfico
```
